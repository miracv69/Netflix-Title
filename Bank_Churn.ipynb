```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Veri setleri/Churn_Modelling.csv")
```

```python
# dropping useless columns
df = df.drop(columns=["RowNumber","CustomerId","Surname"])
```

```python
# Transforming object columns to integer
ohe = OneHotEncoder()
a = ohe.fit_transform(df[["Geography","Gender"]]).toarray()
```

```python
df = df.drop(columns=["Geography","Gender"])
```

```python
df_a = pd.DataFrame()
```

```python
kolon_a = ohe.get_feature_names_out()
```

```python
df_a[kolon_a]=a
```

```python
new_df = pd.concat([df,df_a],axis=1)
```

```python
# Creating the model
y = new_df["Exited"]
x = new_df.drop(columns=["Exited"])
```

```python
xt,xtest,yt,ytest=train_test_split(x,y,random_state=61,train_size=0.7)
```

```python
rf = RandomForestClassifier(n_estimators=400,max_depth=5)
```

```python
model = rf.fit(xt,yt)
```

```python
geography_gender = ohe.transform([["Spain","Male"]]).toarray().tolist()[0]
```

```python
prediction = [600,20,2,25000,2,1,1,10000]
```

```python
prediction.extend(geography_gender)
```

```python
predict = pd.DataFrame([prediction],columns=x.columns)
```

```python
# Prediction
model.predict(predict)
```
